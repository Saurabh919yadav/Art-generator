```
# Importing necessary librariries
import cv2 # for open-cv
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
from skimage.measure import compare_ssim
```

```
# Reading Image 
img0 = cv2.imread('handwriting.jpg') 
```

```
cv2_imshow(img0)
```

```
# Converting RGB image to Gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
```

### Using Gaussan blur to smoothen image
A Gaussian blur (also known as Gaussian smoothing) is the result of blurring an image by a Gaussian function. It is a widely used effect in graphics software, typically to reduce image noise and reduce detail.

```

img = cv2.GaussianBlur(gray,(3,3),0)
```

### Using Edge detection techniques
Here, we have applied laplacian edge detection and sobel edge detection

```
laplacian = cv2.Laplacian(img,cv2.CV_64F) # Laplacian edge detection
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # Sobel edge detection for horizontal edges
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # Sobel edge detection for vertical edges

#plotting 
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
(score, diff) = compare_ssim(img, laplacian, full=True)
print("SSIM: {}".format(score))

print("Laplacian : ")
cv2_imshow(laplacian)
```

```
(score, diff) = compare_ssim(img, sobelx, full=True)
print("SSIM: {}".format(score))

print('Sobel X :')
cv2_imshow(sobelx)
```

```
(score, diff) = compare_ssim(img, sobely, full=True)
print("SSIM: {}".format(score))
print('Sobel Y : ')
cv2_imshow(sobely)
```
